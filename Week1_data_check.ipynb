```python
import pandas as pd
```

```python
# Step 1: Load data
df = pd.read_csv("C:/Project2_Data/raw_prices.csv", parse_dates=["Date"], index_col="Date")
```

```python
# Step 2: Basic checks
print(df.head())
print(df.info())
```

```python
# Step 3: Check missing values
print("Missing values:\n", df.isnull().sum())
```

```python
# Step 4: Forward fill (safety)
df = df.fillna(method="ffill")
```

```python
# Step 5: Save cleaned data
df.to_csv("clean_prices.csv")
```
