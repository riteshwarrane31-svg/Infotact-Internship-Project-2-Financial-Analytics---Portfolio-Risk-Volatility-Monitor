```python
import pandas as pd
import numpy as np
```

```python
# Load returns
returns = pd.read_csv("log_returns.csv", index_col="Date", parse_dates=True)
```

```python
# Step 1: Portfolio returns
weights = np.array([1 / returns.shape[1]] * returns.shape[1])
portfolio_returns = returns @ weights
```

```python
# Step 2: Annual return & volatility
annual_return = portfolio_returns.mean() * 252
annual_vol = portfolio_returns.std() * np.sqrt(252)
```

```python
# Step 3: Sharpe Ratio
risk_free_rate = 0.05
sharpe_ratio = (annual_return - risk_free_rate) / annual_vol
```

```python
print("Annual Return:", annual_return)
print("Annual Volatility:", annual_vol)
print("Sharpe Ratio:", sharpe_ratio)
```
