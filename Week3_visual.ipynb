```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Load returns
returns = pd.read_csv("log_returns.csv", index_col="Date", parse_dates=True)
```

```python
# Step 1: Rolling volatility (30 days)
rolling_vol = returns.rolling(30).std() * np.sqrt(252)
```

```python
# Step 2: Plot rolling volatility
rolling_vol.plot(figsize=(10, 6))
plt.title("30-Day Rolling Volatility")
plt.xlabel("Date")
plt.ylabel("Volatility")
plt.show()
```

```python
# Step 3: Save for Tableau
rolling_vol.to_csv("rolling_volatility.csv")
```
