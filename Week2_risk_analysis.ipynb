```python
import pandas as pd
import numpy as np
```

```python
# Load clean data
prices = pd.read_csv("clean_prices.csv", index_col="Date", parse_dates=True)
```

```python
# Step 1: Log returns
log_returns = np.log(prices / prices.shift(1)).dropna()
```

```python
# Step 2: Equal portfolio weights
num_assets = len(prices.columns)
weights = np.array([1 / num_assets] * num_assets)
```

```python
# Step 3: Covariance matrix (annualized)
cov_matrix = log_returns.cov() * 252
```

```python
# Step 4: Portfolio volatility
portfolio_volatility = np.sqrt(weights.T @ cov_matrix @ weights)
print("Annual Portfolio Volatility:", portfolio_volatility)
```

```python
# Step 5: Monte Carlo Simulation
num_simulations = 10000
num_days = 252
mean_returns = log_returns.mean()

simulated_returns = []

for _ in range(num_simulations):
    daily_returns = np.random.multivariate_normal(
        mean_returns, cov_matrix / 252, num_days
    )
    portfolio_daily = daily_returns @ weights
    simulated_returns.append(portfolio_daily.sum())
```

```python
# Step 6: Value at Risk (95%)
VaR_95 = np.percentile(simulated_returns, 5)
print("Value at Risk (95%):", VaR_95)
```

```python
# Save outputs
pd.DataFrame(simulated_returns, columns=["Simulated_Returns"]) \
    .to_csv("monte_carlo_returns.csv", index=False)
```
